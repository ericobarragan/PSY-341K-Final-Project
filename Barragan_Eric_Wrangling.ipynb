```python
import numpy as np
import pandas as pd
```

```python
walk = pd.read_csv('data/EPA_Walkability.csv')
air = pd.read_csv('data/annual_aqi_by_county_2019.csv')
```

```python
air
```

```python
#removing missing data 
walk = walk.replace(-99999.00, np.nan)
walk = walk.replace(-99999, np.nan)
walk.isna().sum()
```

```python
walk[['STATEFP','COUNTYFP']] = walk[['STATEFP','COUNTYFP']].astype(str)
walk.loc[:, 'STATEFP'] = walk['STATEFP'].apply(lambda x: str(x).zfill(2))
walk.loc[:, 'COUNTYFP'] = walk['COUNTYFP'].apply(lambda x: str(x).zfill(3))
walk['COUNTY_FIPS'] = walk['STATEFP'] + walk['COUNTYFP']
```

```python
my_list = []
with open('data/fips.txt', 'r') as file:
    for line in file:
        # Process each line as needed
        counties = line.strip().split(maxsplit = 1)
        my_list.append(counties)
        
states = []
with open('data/state_fips.txt', 'r') as file:
    for line in file:
        # Process each line as needed
        state = line.strip().split(maxsplit = 1)
        states.append(state)
```

```python
# Create a mapping dictionary
mapping_dict = dict(my_list)
md = dict(states)

# Use map to convert COUNTY FIPS to COUNTY NAME
walk['County'] = walk['COUNTY_FIPS'].map(mapping_dict).fillna(walk['COUNTY_FIPS'])
walk['State'] = (walk['STATEFP'].map(md).fillna(walk['STATEFP'])).str.title()
walk
```

```python
num = walk.select_dtypes(include=['int', 'float']).columns
walkavg = walk.groupby(['County','State'])[num].mean().reset_index()
walksum = walk.groupby(['County','State'])[num].sum().reset_index()

walkavg
```

```python
air['County'] = air['County'] + ' County'
```

```python
merged = pd.merge(air, walkavg, on = ['County', 'State']) 
mergesum = pd.merge(air, walksum, on = ['County', 'State'])
```

```python

merged.drop(columns = ['Year'], inplace = True)
```

```python
merged.drop(columns = ['OBJECTID',
 'GEOID10',
 'GEOID20',
 'TRACTCE',
 'BLKGRPCE',
 'CSA',
 'CBSA'], inplace = True)
```

```python
merged.drop(columns = [ '90th Percentile AQI',
 'Median AQI','Days with AQI'], inplace = True)
```

```python
mergesum.loc[:, ['TotPop']]
merged['TotPop'] = mergesum['TotPop']
merged['AutoOwn0'] = mergesum['AutoOwn0']
merged['AutoOwn1'] = mergesum['AutoOwn1']
merged['AutoOwn2p']= mergesum['AutoOwn2p']
merged['Ac_Unpr'] = mergesum['Ac_Unpr']
merged['D3A'] = mergesum['D3A']
merged['E8_Ind'] = mergesum['E8_Ind']
```

```python
new = merged.iloc[:,0:14]
```

```python
new = pd.concat([new, merged[['TotPop', 'AutoOwn0', 'AutoOwn1','AutoOwn2p', 'NatWalkInd', 'Ac_Unpr','D3A','E8_Ind']]], axis=1)
```

```python
corr = new.corr(numeric_only = True)

thresh = 0.4

mask = corr.abs().gt(thresh)

x = corr.where(mask).stack()
x.to_csv('data/corr.csv')
```
